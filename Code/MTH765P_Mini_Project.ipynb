```python
'''
    Based on dataframe description. 
    We observe that following column need to be transformed in order to modify scientific notation.
        1. ALL_OTHER_PAY
        2. TOTAL_PAY
'''
#Convert column to numeric
LACity_Payroll_df['ALL_OTHER_PAY'] = pd.to_numeric(LACity_Payroll_df['ALL_OTHER_PAY'], errors='coerce') 
#Transforming each numeric value to string two decimal
LACity_Payroll_df['ALL_OTHER_PAY'] = LACity_Payroll_df['ALL_OTHER_PAY'].apply(lambda x: '{:.2f}'.format(x))
#Modifying the column back to float
LACity_Payroll_df['ALL_OTHER_PAY'] = LACity_Payroll_df['ALL_OTHER_PAY'].astype(float)

LACity_Payroll_df['TOTAL_PAY'] = pd.to_numeric(LACity_Payroll_df['TOTAL_PAY'], errors='coerce')
LACity_Payroll_df['TOTAL_PAY'] = LACity_Payroll_df['TOTAL_PAY'].apply(lambda x: '{:.2f}'.format(x))
LACity_Payroll_df['TOTAL_PAY'] = LACity_Payroll_df['TOTAL_PAY'].astype(float)

#Transforming each numeric value to string two decimal & modifying column dtype back to float
LACity_Payroll_df['OVERTIME_PAY'] = LACity_Payroll_df['OVERTIME_PAY'].apply(lambda x: '{:.2f}'.format(x))
LACity_Payroll_df['OVERTIME_PAY'] = LACity_Payroll_df['OVERTIME_PAY'].astype(float)

#Transforming each numeric value to string two decimal & modifying column dtype back to float
LACity_Payroll_df['REGULAR_PAY'] = LACity_Payroll_df['REGULAR_PAY'].apply(lambda x: '{:.2f}'.format(x))
LACity_Payroll_df['REGULAR_PAY'] = LACity_Payroll_df['REGULAR_PAY'].astype(float)

#Transforming each numeric value to string two decimal & modifying column dtype back to float
LACity_Payroll_df['CITY_RETIREMENT_CONTRIBUTIONS'] = LACity_Payroll_df['CITY_RETIREMENT_CONTRIBUTIONS'].apply(lambda x: '{:.2f}'.format(x))
LACity_Payroll_df['CITY_RETIREMENT_CONTRIBUTIONS'] = LACity_Payroll_df['CITY_RETIREMENT_CONTRIBUTIONS'].astype(float)

#Transforming each numeric value to string two decimal & modifying column dtype back to float
LACity_Payroll_df['BENEFIT_PAY'] = LACity_Payroll_df['BENEFIT_PAY'].apply(lambda x: '{:.2f}'.format(x))
LACity_Payroll_df['BENEFIT_PAY'] = LACity_Payroll_df['BENEFIT_PAY'].astype(float)
```

```python
#Checking for null values in the dataframe
LACity_Payroll_df.isnull().sum()
```

```python
'''
    Based on the above output, we observe that there are null values present for the below mentioned columns:
    1. LAST_NAME
    2. JOB_CLASS_PGRADE
    3. JOB_TITLE
    4. MOU
    5. MOUT_TITLE
    6. GENDER
    7. ETHNICITY
'''
LACity_Payroll_df['ETHNICITY'].fillna('UNKNOWN',inplace = True)
LACity_Payroll_df['JOB_CLASS_PGRADE'].fillna(0,inplace = True)
LACity_Payroll_df['JOB_TITLE'].fillna('MISSING INPUT',inplace = True)
LACity_Payroll_df['GENDER'].fillna('INVALID INPUT', inplace = True)
LACity_Payroll_df['LAST_NAME'].fillna('MISSING INPUT',inplace = True)
LACity_Payroll_df['MOU'].fillna('MISSING INPUT',inplace = True)
LACity_Payroll_df['MOU_TITLE'].fillna('MISSING INPUT',inplace = True)
```

```python
'''
    Standardising values of Gender & MOU_TITLE.
    Changing junk values of Gender to INVALID INPUT. Junk values present:
        1. UNKNOWN 
        2. OTHER
        
    Changing junk values of MOUT_TITLE to -1. Junk values present:
        1. B
        2. P
        3. M
        4. V
        5. W
        6. Z
'''
LACity_Payroll_df['GENDER'].replace(['UNKNOWN','OTHER'],'INVALID INPUT',inplace = True)
LACity_Payroll_df['MOU'].replace(['B','P','M','V','W','Z','U'],-1,inplace = True)
```

```python
'''
    Implemnting scatter plot to analyse relationship between PAY_YEAR vs REGULAR_PAY & TOTAL_PAY
'''
plt.figure(figsize = (8,3))
plt.scatter(LACity_Payroll_df['PAY_YEAR'], LACity_Payroll_df['REGULAR_PAY'])
plt.xlabel('PAY_YEAR')
plt.ylabel('REGULAR_PAY')
plt.show()

plt.figure(figsize = (8,3))
plt.scatter(LACity_Payroll_df['PAY_YEAR'], LACity_Payroll_df['TOTAL_PAY'])
plt.xlabel('PAY_YEAR')
plt.ylabel('TOTAL_PAY')
plt.show()
```

```python
'''
    Creating a new dataframe for the below mentioned filters, to analysis salary accordingly.
    Filters:
        1. Department Title = Police
        2. Job Status = Active
        3. Employment Type = Full Time
        4. Gender = Male / Female / INVALID INPUT
        5. Pay Year = 2019 - 2022
'''

Police_Active_df = LACity_Payroll_df[(LACity_Payroll_df['DEPARTMENT_TITLE']=='POLICE') & (LACity_Payroll_df['JOB_STATUS']=='ACTIVE') & (LACity_Payroll_df['EMPLOYMENT_TYPE']=='FULL_TIME')]

gender_order = ['MALE', 'FEMALE', 'INVALID INPUT']
fig, axes = plt.subplots(1, 6, figsize=(20, 5), sharey=True)
for i, year in enumerate([2018,2019,2020,2021,2022, 2023]):
    ax = sns.barplot(x='GENDER', y='TOTAL_PAY',data=Police_Active_df[Police_Active_df['PAY_YEAR']==year],estimator = sum,ax = axes[i], order=gender_order)
    ax.set_title(f'Total Pay by Gender - {year}')
    ax.set_xlabel('Gender')
    ax.set_ylabel('Total Pay')  
plt.tight_layout()
plt.show()
#print(Police_Active_df.shape)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5),sharey = True)

for i, gender in enumerate(['MALE','FEMALE','INVALID INPUT']):
    ax = sns.lineplot(x='PAY_YEAR',y='TOTAL_PAY',data = Police_Active_df[Police_Active_df['GENDER'] == gender], estimator=sum, ax=axes[i], sort=False, markers = True)
    ax.set_title(f'Total Pay Per Year - {gender}')
    ax.set_xlabel('Gender')
    ax.set_ylabel('Total Pay')
    
plt.tight_layout()
plt.show()
```

```python
# Assuming sorted_ethnicities is defined
# If not, you can sort the unique ethnicities within the specified DataFrame
sorted_ethnicities = sorted(Police_Active_df['ETHNICITY'].unique())

# Filter DataFrame for the year 2022
filtered_df_2022 = Police_Active_df[Police_Active_df['PAY_YEAR'] == 2022]

# Define custom colors for 'MALE' and 'FEMALE'
custom_palette = {'MALE': 'red', 'FEMALE': 'darkblue', 'INVALID INPUT': 'yellow'}

# Create a bar plot with 'GENDER' as the hue and custom colors
plt.figure(figsize=(10, 5))
sns.barplot(x='ETHNICITY', y='REGULAR_PAY', hue='GENDER', data=filtered_df_2022, estimator=sum, order=sorted_ethnicities, palette=custom_palette)
plt.title('Total Pay by Ethnicity - 2022')
plt.xlabel('Ethnicity')
plt.ylabel('Total Pay')
plt.xticks(rotation=90)
plt.legend(title='Gender')
plt.show()
```
