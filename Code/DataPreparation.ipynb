```python
#Merging dataframe to create a single dataset
data_merge1 = pd.merge(races,results,how = 'inner',on = 'raceId')
data_merge2 = pd.merge(data_merge1,qualifying, how = 'inner', on = ['raceId','driverId','constructorId'])
data_merge3 = pd.merge(data_merge2,drivers, how = 'inner', on ='driverId')
data_merge4 = pd.merge(data_merge3,lap_times, how = 'inner', on = ['raceId','driverId'])
data_merge5 = pd.merge(data_merge4,constructor,how='inner',on=['constructorId'])
data_merge6 = pd.merge(data_merge5,constructor_standings, how = 'inner',on = ['raceId', 'constructorId'])
data_merge7 = pd.merge(data_merge6,circuits, how ='inner', on = 'circuitId')
data_merge8 = pd.merge(data_merge7,pit_stops, how = 'outer', on = ['raceId','driverId','lap'])
```

```python
data_merge8.head()
```

```python
#Check if there are any null records after merging with laptimes
filtered_df = data_merge8[data_merge8['resultId'].isnull() | (data_merge8['resultId'] == '')]
```

```python
#Filtering data only from year 2011 till 2020.
finalData = data_merge8[(data_merge8['race_year'] > 2010) & (data_merge8['race_year'] <= 2020)]
```
